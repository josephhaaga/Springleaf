```python
import pandas as pd;
import numpy as np;
```

```python
r = pd.read_csv('train.csv')
```

```python
r
```

```python
a = r.describe()
```

```python
a = a.drop(['ID'],axis=1)
```

```python
sortedFeatures = a.T.sort('min',ascending=False)
```

```python
top50Features = sortedFeatures.index[0:49]
```

```python
top50Subset = r[top50Features]
```

```python
top50Subset
```

```python
top50Features
```

```python
from sklearn.ensemble import BaggingClassifier;
from sklearn.neighbors import KNeighborsClassifier;
bagging = BaggingClassifier(KNeighborsClassifier(),max_samples=0.5,max_features=0.5)
```

```python
top50Subset = top50Subset.fillna(value=0)
```

```python
fitBags = bagging.fit(top50Subset[:140000],r.target[:140000])
```

```python
output = fitBags.predict(top50Subset[140001:])
```

```python
results = [r.target[z]==output[z-140000] for z in range(140001,145230)]
```

```python
print("BaggingClassifier results:");
print(results.count(True)/len(results));
```

```python
from sklearn.ensemble import RandomForestClassifier
X = top50Subset[:140000];
Y = r.target[:140000];
clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X, Y)
```

```python
randomForestOutput = clf.predict(top50Subset[140001:])
randomForestResults = [r.target[z]==randomForestOutput[z-140000] for z in range(140001,145230)]
print("RandomForest results:");
print(randomForestResults.count(True)/len(results));
```
