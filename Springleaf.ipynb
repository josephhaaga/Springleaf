```python
import pandas as pd;
import numpy as np;
```

```python
r = pd.read_csv('train.csv').dropna(how='any',axis=1)
```

```python
a = r.describe()
```

```python
a = a.drop(['ID'],axis=1)
```

```python
sortedFeatures = a.T.sort('min',ascending=False)
```

```python
top50Features = sortedFeatures.index[0:49]
```

```python
top50Subset = r[top50Features]
```

```python
top50Subset
```

```python
top50Features
```

```python
from sklearn.ensemble import BaggingClassifier;
from sklearn.neighbors import KNeighborsClassifier;
bagging = BaggingClassifier(KNeighborsClassifier(),max_samples=0.5,max_features=0.5)
```

```python
top50Subset = top50Subset.fillna(value=0)
```

```python
fitBags = bagging.fit(top50Subset[:140000],r.target[:140000])
```

```python
output = fitBags.predict(top50Subset[140001:])
```

```python
results = [r.target[z]==output[z-140000] for z in range(140001,145230)]
```

```python
print("BaggingClassifier results:");
print(results.count(True)/len(results));
```

```python
from sklearn.ensemble import RandomForestClassifier
X = top50Subset[:140000];
Y = r.target[:140000];
clf = RandomForestClassifier(n_estimators=10)
clf = clf.fit(X, Y)
```

```python
randomForestOutput = clf.predict(top50Subset[140001:])
randomForestResults = [r.target[z]==randomForestOutput[z-140000] for z in range(140001,145230)]
print("RandomForest results:");
print(randomForestResults.count(True)/len(results));
```

```python
Y = r.target
X = r.drop(['target'],axis=1)
X.set_index(X.ID).drop(['ID'],axis=1)
from sklearn.cross_validation import train_test_split
train, test = train_test_split(X, test_size=0.2)
```

```python
train_labels = [Y[z] for z in train.index]
test_labels = [Y[z] for z in test.index]
```

```python
from sklearn import preprocessing

train = train.fillna(value=0)
test = test.fillna(value=0)

le = preprocessing.LabelEncoder()

# get categoricals
# type(train.VAR_0001[1])

isCategorical = [(type(train[z][1])!=np.float64 and type(train[z][1])!=np.int64) for z in train.columns]

# to convert back
# train.Sex = le_sex.inverse_transform(train.Sex)
# isCategorical
```

```python
listOfCategoricals = np.where(isCategorical)[0].tolist()

processed_train = pd.DataFrame();
processed_test = pd.DataFrame();

le.fit_transform(train['VAR_0001'])
```

```python
categoricalVars = train[listOfCategoricals].columns
for c in categoricalVars:
    try:
        processed_train[c] = le.fit_transform(train[c].fillna(value=0))
        processed_test[c] = le.fit_transform(test[c].fillna(value=0))
    except:
        pass;
    
train_copy = train.copy(deep=True)
test_copy = test.copy(deep=True)
for c in categoricalVars:
    try:
        train_copy[c] = processed_train[c]
        test_copy[c] = processed_test[c]
    except:
        train_copy.drop([c],axis=1)
        test_copy.drop([c],axis=1)

train_copy
```
